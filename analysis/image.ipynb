```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = '15'
plt.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
```

```python

im1 = Image.open('../phantoms/0323_focus_point.tif')
im2 = Image.open('../phantoms/0321_focus_wrap_point.tif')

Nx = 216         # Number of grid points
Ny = 236
size = 100
dx = 100/Nx

x = np.linspace(0, 1, Nx ) * size 
y = np.linspace(-1, 1, Ny) * size/2 *Ny/Nx

xnew = np.linspace(0, 1, 256 ) * size 
ynew = np.linspace(-1, 1, 256) * size/2 *Ny/Nx

#point = np.pad(point, [(20,20), (10,10)], mode='constant', constant_values=1540)

im1_arr = np.array(im1, dtype=np.float64)
im2_arr = np.array(im2, dtype=np.float64)

fig, ax = plt.subplots(1,2, figsize=(11,4))
im = ax[0].pcolormesh(ynew, xnew, im1_arr)
ax[1].pcolormesh(ynew, xnew, im2_arr )

# Plot 
plt.set_cmap('gray')
ax[0].invert_yaxis()
ax[1].invert_yaxis()
fig.colorbar(im, ax=ax)

ax[0].set_title('Unwrapped')
ax[1].set_title('Wrapped')

ax[0].set_ylabel('Axial Coordinate (mm)')
fig.text(0.43, 0.01, 'Lateral Coordinate (mm)', ha='center')

plt.savefig('point.png', dpi=300, bbox_inches='tight')
```

```python
spheres = np.loadtxt('../phantoms/spheres.csv', delimiter=',')
point = np.loadtxt('../phantoms/point.csv', delimiter=',')
anechoic = np.loadtxt('../phantoms/anechoic.csv', delimiter=',')

plt.set_cmap('gray')
fig, ax = plt.subplots(1,3, figsize=(16,4))

im = ax[0].pcolormesh(y,x,  point, vmin=1400, vmax=1600)
im = ax[1].pcolormesh(y,x,  spheres, vmin=1400, vmax=1600)
im = ax[2].pcolormesh(y,x,  anechoic, vmin=1400, vmax=1600)
ax[0].title.set_text('1. Point')
ax[1].title.set_text('2. Ball')
ax[2].title.set_text('3. Anechoic Chamber')
ax[0].invert_yaxis()
ax[1].invert_yaxis()
ax[2].invert_yaxis()

ax[0].set_ylabel('Axial Coordinate (mm)')
fig.text(0.43, 0.01, 'Lateral Coordinate (mm)', ha='center')

cbar = plt.colorbar(im, ax = ax)

plt.savefig('phantoms.png', dpi=300, bbox_inches='tight')
```
